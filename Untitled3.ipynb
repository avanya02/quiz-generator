```python
import re
import pandas as pd

# Load your CSV file (replace 'file_path' with the actual path)
data = pd.read_csv('training-data.csv')

# Extract the text from the loaded CSV's column, replacing NaN with empty strings
raw_text = data.iloc[:, 0].fillna('')  # assuming all text is in the first column

# Initialize lists to store structured data
titles = []
descriptions = []

# A regex pattern to identify potential titles (e.g., questions, headings)
title_pattern = re.compile(r'^[A-Z].*\?$')  # Assuming titles end with a question mark

# Loop through the text data and structure it
current_title = None
current_description = []

for line in raw_text:
    # Ensure the line is a string before applying regex
    if isinstance(line, str) and re.match(title_pattern, line):
        # If a title is found, store the current title and description (if any)
        if current_title and current_description:
            titles.append(current_title)
            descriptions.append(" ".join(current_description))

        # Reset for the new title
        current_title = line
        current_description = []
    else:
        # Otherwise, treat it as part of the description
        current_description.append(line)

# Add the last title-description pair
if current_title and current_description:
    titles.append(current_title)
    descriptions.append(" ".join(current_description))

# Create a structured DataFrame
structured_data = pd.DataFrame({
    'Title': titles,
    'Description': descriptions
})

# Save to a new CSV file
structured_data.to_csv('structured_data.csv', index=False)

structured_data.head()
```
